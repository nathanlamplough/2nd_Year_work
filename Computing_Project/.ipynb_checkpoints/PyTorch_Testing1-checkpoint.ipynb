```python
import torch 
import torch.nn as nn
import pandas as pd
import numpy as np
from sklearn.preprocessing import normalize


dataset = pd.read_csv('C:/Users/natha/ThirdYear/Computing_Project/data/heart-disease-uci/heart.csv', delimiter = ',', quoting = 3)
Xd = dataset.drop(['target'], axis = 1)
Yd = dataset[['target']]
Xd[:10]
```

```python
n_in, n_h, n_out, batch_size = 13, 15, 1, 10
X = torch.tensor(Xd.values, dtype=torch.float)(Xd - 2)/0.5
Y = torch.tensor(Yd.values, dtype=torch.float)
#X[:10]
```

```python
model = nn.Sequential(nn.Linear(n_in, n_h),
                     nn.ReLU(),
                      nn.Linear(n_h, n_h),
                     nn.ReLU(),
                     nn.Linear(n_h, n_out),
                     nn.Sigmoid())
        
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
for epoch in range(10):
    y_pred = model(X)
    
    loss = criterion(y_pred, Y)

    print('epoch: ', epoch, ' loss: ', loss.item())
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()

    
```
