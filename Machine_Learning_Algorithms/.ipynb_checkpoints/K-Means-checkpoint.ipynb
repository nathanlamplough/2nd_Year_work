```python
import tensorflow as tf
from tensorflow import 
import pandas as pd
import numpy as np

data = pd.read_excel('C:/Users/natha/Desktop/GitRepo/NathanLamplough/Data/CardiologyNumerical.xlsx', delimiter = ',', quoting = 3 )
x = data.drop(['class'], axis = 1)
x[:10]
```

```python
dimensions = len(x.columns)

def input_fn():
    return tf.train.limit_epochs(tf.convert_to_tensor(x,dtype = tf.float32), num_epochs =1)
```

```python
num_clusters = 20
kmeans = tf.contrib.factorization.KMeansClustering(num_clusters=num_clusters, use_mini_batch = False)
```

```python
#training
num_iterations = 100
previous_centres = None

for i in range(num_iterations):
    kmeans.train(input_fn)
    cluster_centres = kmeans.cluster_centres()
    if previous_centres is not None:
        print('delta:', cluster_centres - previous_centres)
    previous_centres = cluster_centres
    print('score:', kmeans.score(input_fn))
print('cluster centres:', cluster_centres)
```
