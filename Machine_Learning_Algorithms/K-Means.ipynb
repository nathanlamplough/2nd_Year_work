```python
import tensorflow as tf
import pandas as pd
import numpy as np

data = pd.read_csv('C:/Users/natha/Desktop/GitRepo/NathanLamplough/Data/Iris.csv', delimiter = ',', quoting = 3 )
x = data.drop(['Species','Id'], axis = 1)
x[:10]
```

```python
dimensions = len(x.columns)
x_data = np.asarray(x)
def input_fn():
    return tf.train.limit_epochs(tf.convert_to_tensor(x_data, dtype = tf.float32), num_epochs =1)
```

```python
num_clusters = 4
kmeans = tf.contrib.factorization.KMeansClustering(num_clusters=num_clusters, use_mini_batch = False)
```

```python
#training
num_iterations = 10
previous_centres = None

for i in range(num_iterations):
    kmeans.train(input_fn)
    cluster_centres = kmeans.cluster_centers()
    if previous_centres is not None:
        print('delta:', cluster_centres - previous_centres)
    previous_centres = cluster_centres
    print('score:', kmeans.score(input_fn))
print('cluster centres:', cluster_centres)
```
