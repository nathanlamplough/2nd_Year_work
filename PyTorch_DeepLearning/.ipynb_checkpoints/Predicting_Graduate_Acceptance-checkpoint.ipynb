### Modular Neural network class 
This will be the method of performing regression and predicting the acceptance rate of the graduates
***
The Class Parameters are as follows
- d_in = the dimensions of the instance vector
- h = the number of hidden nodes
- d_out = the dimensions of the output
- l = the number of layers
- p = the probability hyperparameters for dropout

```python
class Net(nn.Module):
    def __init__(self, d_in, h, d_out, l, p):
        super(Net, self).__init__()
        self.h = h
        self.in_layer = nn.Linear(d_in, h)
        self.layers = []
        self.out_layer = nn.Linear(h, d_out)
        self.drop_layer = nn.Dropout(p=p);
        self.add_layers(l)
        self.l = l
        self.d_out = d_out

    def add_layers(self, l):
        for i in range(l):
            self.layers.append(nn.Linear(self.h, self.h))

    def forward(self, x):
        out = F.relu(self.in_layer(x))
        for i in range(self.l):
            out = self.layers[i](out)
            y_pred = torch.sigmoid(self.out_layer(out))
        return y_pred
        
```
